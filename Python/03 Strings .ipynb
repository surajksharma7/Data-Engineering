#### creating strings 

```python
s = 'hello' 
s = "hello"
s = '''hello'''
s = """hello"""
s = str('hello')
```

#### Accessing substrings from a string

```python
# indexing 
s = "hello world"
```

```python
s[0]
```

```python
s[-1]
```

```python
# slicing [start, stop, step] 
s[0:4] 
```

#### Editing and deleting in strings 

```python
# we cannot edit and delete in strings because it is immutable 
```

```python
s = 'hello world'
del s 
```
